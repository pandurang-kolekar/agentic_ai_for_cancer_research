## Section 1: Setup and Imports

We prepare deterministic foresight analyses for future biomedical AI capabilities.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', context='talk')

# Configuration
plt.rcParams['figure.figsize'] = (12, 6)
np.random.seed(42)

print('Environment configured for Chapter 16 future directions notebook.')
```

## Section 2: Load Data and Explore

We define a roadmap of future capabilities with impact and maturity-gap estimates.

```python
roadmap_df = pd.DataFrame([
    {'capability': 'self_verifying_agents', 'horizon': 'near_term', 'impact': 3, 'maturity_gap': 2},
    {'capability': 'active_learning_experiments', 'horizon': 'near_term', 'impact': 3, 'maturity_gap': 1},
    {'capability': 'multimodal_patient_digital_twins', 'horizon': 'mid_term', 'impact': 5, 'maturity_gap': 4},
    {'capability': 'autonomous_hypothesis_networks', 'horizon': 'mid_term', 'impact': 5, 'maturity_gap': 3},
    {'capability': 'fully_governed_discovery_loops', 'horizon': 'long_term', 'impact': 5, 'maturity_gap': 5},
])
display(roadmap_df)
```

## Section 3: Main Analysis

We rank roadmap items by expected impact and maturity gap.

```python
roadmap_df['priority_score'] = roadmap_df['impact'] * roadmap_df['maturity_gap']
roadmap_df['readiness'] = np.where(roadmap_df['maturity_gap'] <= 2, 'ready_now', 'needs_infrastructure')
roadmap_summary = roadmap_df.groupby('horizon', as_index=False).agg(avg_priority=('priority_score', 'mean'), capabilities=('capability', 'count'))
roadmap_summary['avg_priority'] = roadmap_summary['avg_priority'].round(2)
display(roadmap_df.sort_values('priority_score', ascending=False))
display(roadmap_summary)
```

## Section 4: Visualization

We visualize capability priorities and readiness gaps.

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.barplot(data=roadmap_df, x='capability', y='priority_score', hue='horizon', ax=axes[0])
axes[0].set_title('Future Capability Priority')
axes[0].set_xlabel('Capability')
axes[0].set_ylabel('Priority Score')
axes[0].tick_params(axis='x', rotation=25)

sns.scatterplot(data=roadmap_df, x='maturity_gap', y='impact', hue='readiness', s=150, ax=axes[1])
axes[1].set_title('Impact vs Maturity Gap')
axes[1].set_xlabel('Maturity Gap')
axes[1].set_ylabel('Impact')

plt.tight_layout()
plt.show()
```

## Section 5: Exercises

```python
priority_df = roadmap_df[['capability', 'priority_score', 'readiness']].sort_values('priority_score', ascending=False)
display(priority_df)
```

### Exercise 2: Scenario Planner

*Description: Draft a short scenario sentence for each capability horizon.*

```python
scenario_df = roadmap_df[['capability', 'horizon', 'maturity_gap']].copy()
scenario_df['scenario'] = scenario_df.apply(lambda row: f"Pilot {row['capability']} for {row['horizon']} planning", axis=1)
display(scenario_df)
```
